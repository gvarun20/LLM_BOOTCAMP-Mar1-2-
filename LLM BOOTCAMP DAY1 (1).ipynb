```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import matplotlib.pyplot as plt

# Assuming you've already read the CSV file
csv_file = pd.read_csv(r"C:\Users\Hi\Dataset1\c2p2.csv")

# Set features and target variables
features = csv_file[['AT', 'V', 'AP', 'RH']]  # Independent variables
target = csv_file['PE']  # Dependent variable

# Split the data into train and test sets with a 75-25 ratio
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.25, random_state=42)

# Normalize the features using Min-Max scaling
scaler = MinMaxScaler()
X_train_normalized = scaler.fit_transform(X_train)
X_test_normalized = scaler.transform(X_test)

# Linear Regression model
linear_model = LinearRegression()
linear_model.fit(X_train_normalized, y_train)
y_train_linear = linear_model.predict(X_train_normalized)
y_test_linear = linear_model.predict(X_test_normalized)

# Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train_normalized, y_train)
y_train_rf = rf_model.predict(X_train_normalized)
y_test_rf = rf_model.predict(X_test_normalized)

# Neural Network model
nn_model = MLPRegressor(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42)
nn_model.fit(X_train_normalized, y_train)
y_train_nn = nn_model.predict(X_train_normalized)
y_test_nn = nn_model.predict(X_test_normalized)

# Evaluate the models
def evaluate_model(y_true, y_pred, model_name):
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    print(f'{model_name} Metrics:')
    print(f'Mean Squared Error (MSE): {mse:.2f}')
    print(f'Mean Absolute Error (MAE): {mae:.2f}')
    print(f'R-squared (R2): {r2:.2f}')
    print(f'Mean Absolute Percentage Error (MAPE): {mape:.2f}%\n')

# Evaluate and print metrics for each model
evaluate_model(y_train, y_train_linear, 'Linear Regression (Train)')
evaluate_model(y_test, y_test_linear, 'Linear Regression (Test)')

evaluate_model(y_train, y_train_rf, 'Random Forest (Train)')
evaluate_model(y_test, y_test_rf, 'Random Forest (Test)')

evaluate_model(y_train, y_train_nn, 'Neural Network (Train)')
evaluate_model(y_test, y_test_nn, 'Neural Network (Test)')

# Visualize actual vs predicted values for each model
def plot_predictions(y_true, y_pred, model_name):
    plt.scatter(y_true, y_pred)
    plt.xlabel("Actual Power Output")
    plt.ylabel("Predicted Power Output")
    plt.title(f"Actual vs Predicted Power Output - {model_name}")
    plt.show()

# Plot for Linear Regression
plot_predictions(y_train, y_train_linear, 'Linear Regression (Train)')
plot_predictions(y_test, y_test_linear, 'Linear Regression (Test)')

# Plot for Random Forest
plot_predictions(y_train, y_train_rf, 'Random Forest (Train)')
plot_predictions(y_test, y_test_rf, 'Random Forest (Test)')

# Plot for Neural Network
plot_predictions(y_train, y_train_nn, 'Neural Network (Train)')
plot_predictions(y_test, y_test_nn, 'Neural Network (Test)')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Assuming you've already read the CSV file
csv_file = pd.read_csv(r"C:\Users\Hi\Dataset1\c2p2.csv")

# Set features and target variables
features = csv_file[['AT', 'V', 'AP', 'RH']]  # Independent variables
target = csv_file['PE']  # Dependent variable

# Split the data into train and test sets with a 75-25 ratio
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.25, random_state=42)

# Normalize the features using Min-Max scaling
scaler = MinMaxScaler()
X_train_normalized = scaler.fit_transform(X_train)
X_test_normalized = scaler.transform(X_test)

# Linear Regression model
linear_model = LinearRegression()
linear_model.fit(X_train_normalized, y_train)
y_test_linear = linear_model.predict(X_test_normalized)

# Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train_normalized, y_train)
y_test_rf = rf_model.predict(X_test_normalized)

# Neural Network model
nn_model = MLPRegressor(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42)
nn_model.fit(X_train_normalized, y_train)
y_test_nn = nn_model.predict(X_test_normalized)

# Evaluate the models
def evaluate_model(y_true, y_pred, model_name):
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    print(f'{model_name} Metrics:')
    print(f'Mean Squared Error (MSE): {mse:.2f}')
    print(f'Mean Absolute Error (MAE): {mae:.2f}')
    print(f'R-squared (R2): {r2:.2f}')
    print(f'Mean Absolute Percentage Error (MAPE): {mape:.2f}%\n')
    return mse, mae, r2, mape

# Evaluate and print metrics for each model on the test set
linear_metrics = evaluate_model(y_test, y_test_linear, 'Linear Regression (Test)')
rf_metrics = evaluate_model(y_test, y_test_rf, 'Random Forest (Test)')
nn_metrics = evaluate_model(y_test, y_test_nn, 'Neural Network (Test)')

# Compare the models based on MSE, MAE, R2, and MAPE
best_model = min([linear_metrics, rf_metrics, nn_metrics], key=lambda x: (x[0], x[1], -x[2], x[3]))

print(f"The best model is {best_model[0]} with metrics:")
print(f'Mean Squared Error (MSE): {best_model[0]:.2f}')
print(f'Mean Absolute Error (MAE): {best_model[1]:.2f}')
print(f'R-squared (R2): {best_model[2]:.2f}')
print(f'Mean Absolute Percentage Error (MAPE): {best_model[3]:.2f}%')
```
